```
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np
```

```
file = open('Mouse12-120806.txt','r')

positions = []
for line in file:
  line = line.replace('\n', '')
  line = line.split('\t')
  positions.append(line)
```

```
positions[10]
```

```
positions = np.array(positions).astype('float')
```

```
positions[10]
```

```
positions.shape
```

```
positions = positions[3:]
```

```
x = positions[:,0]
y = positions[:,1]

plt.plot(x,y)
```

```
index_x = np.where(x != -1)[0]
positions = positions[index_x, :]
x = positions[:,0]
y = positions[:,1]
plt.plot(x,y)
```

```
index_y = np.where(y != -1)[0]
positions = positions[index_y, :]
x = positions[:,0]
y = positions[:,1]
plt.plot(x,y)
```

```
x = x[::10]
y = y[::10]
```
